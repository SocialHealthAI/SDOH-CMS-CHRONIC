### Extract data

```python
CMSChronic_year = parm_CMSChronic_years.pop(0)

# Import the first excel file and create dataframe
url = 'https://www.cms.gov/Research-Statistics-Data-and-Systems/Statistics-Trends-and-Reports/Chronic-Conditions/Downloads/CC_Prev_State_County_Age.zip'
r = requests.get(url)
with open('./myzip.zip', 'wb') as f:
    f.write(r.content) 

from zipfile import ZipFile  
# loading the temp.zip and creating a zip object
with ZipFile('./myzip.zip', 'r') as zObject:
# Extracting all files
    zObject.extractall("../CMSchronic")
    zObject.close()
  
dfCMSChronic = pd.read_excel('../CMSchronic/County_Table_Chronic_Conditions_Prevalence_by_Age_' + CMSChronic_year + '.xlsx',
                  sheet_name = 'All Beneficiaries', header=4)
dfCMSChronic = dfCMSChronic.drop(dfCMSChronic.index [ [ 0,1,2 ] ])
dfCMSChronic.columns.values[0]='STATE'
dfCMSChronic.columns.values[1]='COUNTY'
dfCMSChronic.columns.values[2]='FIPS'
dfCMSChronic = dfCMSChronic[['STATE', 'COUNTY', 'FIPS', parm_CMSChronic_disease ]].copy()
dfCMSChronic['YEAR'] = CMSChronic_year

# Import the remaining excel files and append to dataframe
for CMSChronic_year in parm_CMSChronic_years:
    url = 'https://www.cms.gov/Research-Statistics-Data-and-Systems/Statistics-Trends-and-Reports/Chronic-Conditions/Downloads/CC_Prev_State_County_Age.zip'
    r = requests.get(url)
    with open('./myzip.zip', 'wb') as f:
        f.write(r.content) 

    from zipfile import ZipFile  
    # loading the temp.zip and creating a zip object
    with ZipFile('./myzip.zip', 'r') as zObject:
    # Extracting all files
        zObject.extractall("../CMSchronic")
        zObject.close()

    dfCMSChronicNext = pd.read_excel('../CMSchronic/County_Table_Chronic_Conditions_Prevalence_by_Age_' + CMSChronic_year + '.xlsx',
                      sheet_name = 'All Beneficiaries', header=4)
    dfCMSChronicNext = dfCMSChronicNext.drop(dfCMSChronicNext.index [ [ 0,1,2 ] ])
    dfCMSChronicNext.columns.values[0]='STATE'
    dfCMSChronicNext.columns.values[1]='COUNTY'
    dfCMSChronicNext.columns.values[2]='FIPS'
    dfCMSChronicNext = dfCMSChronicNext[['STATE', 'COUNTY', 'FIPS', parm_CMSChronic_disease ]].copy()
    dfCMSChronicNext['YEAR'] = CMSChronic_year
    dfCMSChronic = dfCMSChronic.append(dfCMSChronicNext)
```

### Clean data

```python
dfCMSChronic = dfCMSChronic[dfCMSChronic[parm_CMSChronic_disease] != '* ']
dfCMSChronic = dfCMSChronic[dfCMSChronic.COUNTY != '  ']
dfCMSChronic = dfCMSChronic[dfCMSChronic.COUNTY != '']
dfCMSChronic[parm_CMSChronic_disease] = dfCMSChronic[parm_CMSChronic_disease].apply(pd.to_numeric, errors= 'ignore')
dfCMSChronic.columns = dfCMSChronic.columns.str.upper()
dfCMSChronic['STATE'] = dfCMSChronic['STATE'].str.strip()
dfCMSChronic['COUNTY'] = dfCMSChronic['COUNTY'].str.strip()
```
