### Extract data Using Socratic SODA API

See https://dev.socrata.com/consumers/getting-started.html

```python
CDCPlaces_year = parm_CDCPlaces_years.pop(0)
listYearWhere = "year='" + CDCPlaces_year + "'"
for CDCPlaces_year in parm_CDCPlaces_years:
    listYearWhere = listYearWhere + " OR year='" + CDCPlaces_year + "'"

url = "https://chronicdata.cdc.gov/resource/swc5-untb.json?" + \
"$select=" \
"data_value AS " + parm_CDCPlaces_measure_id + " " + \
",year AS YEAR " + \
",statedesc AS STATE " + \
",locationname AS COUNTY " + \
",locationid AS FIPS" + \
"&$where=measureid='" + parm_CDCPlaces_measure_id + "' AND (" + listYearWhere + ")" + \
"&$limit=50000"
#url
r = requests.get(url)
dfCDCPlaces = pd.read_json(r.text)
dfCDCPlaces
```
